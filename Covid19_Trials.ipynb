```python
#Make 5 questions and solve them by using group by
```

```python
# Group 1
data = {'Acronym': ['TAC-COVID19', 'TMF-COVID-19', 'FALP-COVID', 'MindMyMindFU'],
        'ValueColumn': [10, 20, 30, 40]}
df = pd.DataFrame(data)

# Group by 'GroupColumn' and calculate the mean for each group
grouped_df = df.groupby('Acronym')['ValueColumn'].mean()

# Display the result
print(grouped_df)
```

```python
# Group 2
data = {'Status': ['Recruiting', 'Active, not recruiting', 'Not yet recruiting', 'Completed', 'Suspended'],
        'ValueColumn': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Group by 'GroupColumn' and calculate the mean for each group
grouped_df = df.groupby('Status')['ValueColumn'].mean()

# Display the result
print(grouped_df)
```

```python
# Make 5 questions and solve them by using conditional statements
```

```python
#conditional statements 1- #how many rows had Recruiting
df[df['Status']=='Recruiting'] 
```
